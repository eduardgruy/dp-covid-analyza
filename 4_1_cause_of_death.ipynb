```python
data = (pd.read_csv("./umrtia_mesiace.csv",header=[0,1])
                    .set_index('diagnoza'))

data.columns = ['-'.join(col).strip() for col in data.columns.values]

data = data.transpose().replace('.',0)
data.index = pd.to_datetime(data.index)
```

```python
data.head()
```

```python
data.mean().sort_values(ascending=False).head(3)
```

```python
top_3 = data.mean().sort_values(ascending=False).head(3).index.tolist()
covid = data.transpose().tail(1).index.tolist()
data[top_3 + covid].sort_index().plot_bokeh(
            figsize=(800,600),
         legend=True
 )
```

### umrtnost na dychacie choroby

```python
## umrtnost na dychacie choroby

priemer_2009_19 = data[top_3[2]].resample('1Y').sum().loc['2009-12-31' :'2019-12-31 '].mean()
umrtia_2021 = data[top_3[2]].resample('1Y').sum().loc['2021-12-31']

(umrtia_2021 / priemer_2009_19)
```

```python
## standardna ochylka od priemeru

(umrtia_2021 - priemer_2009_19) / data[top_3[2]].resample('1Y').sum().loc['2009-12-31':'2019-12-31 '].std()
```

### umrtnost na nadory

```python
## rozdiel úmrtí 2021 vs priemer

(umrtia_2021 - priemer_2009_19)
```

```python
## umrtnost na nadory

nd_priemer_2009_19 = data[top_3[1]].resample('1Y').sum().loc['2009-12-31' :'2019-12-31 '].mean()
nd_umrtia_2021 = data[top_3[1]].resample('1Y').sum().loc['2021-12-31']

(nd_umrtia_2021 / nd_priemer_2009_19)
```

### Zmena nádory priemer vs Január 2022

```python
data[top_3[1]].loc['2021-01-01']
```

```python
nd_feb_priemer_2016_19 = data[data.index.month == 2][top_3[1]].sort_index().loc['2009-02-01':'2019-02-01'].mean()
nd_feb_umrtia_2021 = data[top_3[1]].loc['2021-02-01']

(nd_feb_umrtia_2021 / nd_feb_priemer_2016_19)
```

```python
(nd_feb_umrtia_2021 - nd_feb_priemer_2016_19) / data[data.index.month == 2][top_3[1]].sort_index().loc['2009-02-01':'2019-02-01'].std()
```
